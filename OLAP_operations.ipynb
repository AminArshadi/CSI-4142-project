# Part 1: Standard OLAP operations

```python
# Drill down - retrieves details of product categories and subcategories
result = db.aggregated_sales_fact_table.find({}, {'product_category': 1, 'product_sub_category': 1, 'total_sales_amount': 1})

result = pd.DataFrame(list(result))

result.drop('_id', axis=1, inplace=True)

result
```

```python
# Roll up - aggregates sales by product category
result = db.aggregated_sales_fact_table.aggregate([{'$group': {'_id': '$product_category', 'total_sales': {'$sum': '$total_sales_amount'}}}])

result = pd.DataFrame(list(result))

result.rename(columns={'_id': 'product_category'}, inplace=True)

result
```

```python
# Slice - retrieves all sales for the 'electronics' category
result = db.aggregated_sales_fact_table.find({'product_category': 'electronics'})

result = pd.DataFrame(list(result))

result.drop('_id', axis=1, inplace=True)

result
```

```python
# Dice - selecting 'electronics' in January
result = db.aggregated_sales_fact_table.find({'product_category': 'electronics', 'month': 1})

result = pd.DataFrame(list(result))

result.drop('_id', axis=1, inplace=True)

result
```

```python
# Dice - selecting 'hobbies' with more than 5 reviews
result = db.aggregated_product_reviews_fact_table.find({'product_category': 'hobbies', 'review_count': {'$gte': 5}})

result = pd.DataFrame(list(result))

result.drop('_id', axis=1, inplace=True)

result
```

```python
# Combining OLAP operations - slicing the data by 'hobbies' category and then rolling up by months to see the sales distribution across months
result = db.aggregated_sales_fact_table.aggregate([{'$match': {'product_category': 'hobbies'}}, {'$group': {'_id': '$month', 'total_sales': {'$sum': '$total_sales_amount'}}}])

result = pd.DataFrame(list(result))

result.rename(columns={'_id': 'month'}, inplace=True)

result = result.sort_values(by='month', ascending=True).reset_index().drop('index', axis=1)

result
```

```python
# Combining OLAP operations - the average review count for all hobby products within the 'model trains & railway sets' subcategory sold in January, regardless of the year
result = db.aggregated_product_reviews_fact_table.aggregate([
    {'$match': {'product_category': 'hobbies', 'product_sub_category': 'model trains & railway sets'}},
    {'$lookup': {
        'from': 'aggregated_sales_fact_table',
        'localField': 'product_id',
        'foreignField': 'product_id',
        'as': 'sales_data'
    }},
    {'$unwind': '$sales_data'},
    {'$match': {'sales_data.month': 1}},
    {'$group': {
        '_id': '$product_sub_category',
        'average_review_count': {'$avg': '$review_count'}
    }}
])

result = pd.DataFrame(list(result))

result.rename(columns={'_id': 'product_sub_category'}, inplace=True)

result
```

```python
# Combining OLAP operations -  the total sales and average product price for all 'games' products, aggregated by the product subcategory, but only for the year 2015
result = db.aggregated_sales_fact_table.aggregate([
    {'$match': {'product_category': 'games', 'year': 2015}},
    {'$group': {
        '_id': '$product_sub_category',
        'total_sales': {'$sum': '$total_sales_amount'},
        'average_product_price': {'$avg': '$product_price'}
    }}
])

result = pd.DataFrame(list(result))

result.rename(columns={'_id': 'product_sub_category'}, inplace=True)

result
```

```python
# Combining OLAP operations - average monthly sales for electronics sold by the Redmi company
result = db.aggregated_sales_fact_table.aggregate([
{
    "$match":
      {
        "product_company": "Redmi",
        "product_category": "electronics"
      },
  },
  {
    "$group":
      {
        "_id": "$month",
        "sales": {
          "$avg": "$quantity_sold",
        },
      },
  },
  {
    "$project":
      {
        "sales": "$sales",
      },
  },
  {
    "$sort":
      {
        "_id": 1,
      },
  },
])

result = pd.DataFrame(list(result))

result.rename(columns={'_id': 'month'}, inplace=True)

result
```

# Part 2: Explorative operation

```python
# Iceberg - finding the top 5 products with the highest sales amounts
result = db.aggregated_sales_fact_table.aggregate([{'$group': {'_id': '$product_name', 'total_sales': {'$sum': '$total_sales_amount'}}}, {'$sort': {'total_sales': -1}}, {'$limit': 5}])

result = pd.DataFrame(list(result))

result.rename(columns={'_id': 'product_name'}, inplace=True)

result
```

```python
# Windowing - for each product in the 'games' category, calculate the moving average of sales for the past three months on a monthly basis
result = db.aggregated_sales_fact_table.aggregate([
    {'$match': {'product_category': 'games'}},
    {'$sort': {'product_id': 1, 'year': 1, 'month': 1}},
    {'$group': {
        '_id': {
            'product_id': '$product_id',
            'year': '$year',
            'month': '$month'
            },
        'sales_amounts': {'$push': '$total_sales_amount'}
      }
    },
    {
        '$lookup': {
            'from': 'aggregated_sales_fact_table',
            'as': 'preceding_sales',
            'let': {
                'product_id': '$_id.product_id',
                'year': '$_id.year',
                'month': '$_id.month'
            },
            'pipeline': [
                {'$match':
                    {'$expr':
                        {'$and': [
                            {'$eq': ['$product_category', 'electronics']},
                            {'$eq': ['$product_id', '$$product_id']},
                        ]}
                    }
                },
                {'$project': {'total_sales_amount': 1, 'month': 1}}
            ]
        }
    },
    {
        '$project': {
            'product_id': '$_id.product_id',
            'year': '$_id.year',
            'month': '$_id.month',
            'moving_average': {
                '$avg': {
                    '$concatArrays': ['$sales_amounts', '$preceding_sales.total_sales_amount']
                }
            }
        }
    }
])

result = pd.DataFrame(list(result))

result.drop('_id', axis=1, inplace=True)

result
```

```python
# Windowing - calculating the running total of sales amount for 'hobbies' products, sorted by date within each product
result = db.aggregated_sales_fact_table.aggregate([
    {'$match': {'product_category': 'hobbies'}},
    {'$sort': {'product_id': 1, 'year': 1, 'month': 1, 'day': 1}},
    {
        '$lookup': {
            'from': 'aggregated_sales_fact_table',
            'let': {'current_sale_id': '$sale_id', 'current_product_id': '$product_id'},
            'pipeline': [
                {
                    '$match': {
                        '$expr': {
                            '$and': [
                                {'$eq': ['$product_id', '$$current_product_id']},
                                {'$lte': ['$sale_id', '$$current_sale_id']}
                            ]
                        }
                    }
                },
                {
                    '$group': {
                        '_id': '$product_id',
                        'running_total': {'$sum': '$total_sales_amount'}
                    }
                }
            ],
            'as': 'running_totals'
        }
    },
    {'$unwind': '$running_totals'},
    {'$project': {
        'product_id': 1,
        'sale_id': 1,
        'total_sales_amount': 1,
        'running_total': '$running_totals.running_total'
    }}
])

result = pd.DataFrame(list(result))

result.drop('_id', axis=1, inplace=True)

result
```
